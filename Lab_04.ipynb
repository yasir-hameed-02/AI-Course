```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age


class Job:
    def __init__(self, designation, salary):
        self.designation = designation
        self.salary = salary


class Employee(Person, Job):
    def __init__(self, name, age, designation, salary):
        Person.__init__(self, name, age)
        Job.__init__(self, designation, salary)

    def display_details(self):
        print("\n--- Employee Details ---")
        print("Name:", self.name)
        print("Age:", self.age)
        print("Designation:", self.designation)
        print("Salary:", self.salary)


# ----- User Input -----
name = input("Enter employee name: ")
age = int(input("Enter age: "))
designation = input("Enter designation: ")
salary = float(input("Enter salary: "))

emp = Employee(name, age, designation, salary)
emp.display_details()
```

```python
import math

class Circle:
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * self.radius ** 2

    def perimeter(self):
        return 2 * math.pi * self.radius


# ----- User Input -----
r = float(input("Enter radius of the circle: "))
c = Circle(r)
print("\nArea of circle =", c.area())
print("Perimeter of circle =", c.perimeter())
```

```python
class Power:
    def pow(self, x, n):
        result = 1
        if n > 0:
            for i in range(n):
                result *= x
        elif n < 0:
            for i in range(-n):
                result *= x
            result = 1 / result
        return result


# ----- User Input -----
x = float(input("Enter base number: "))
n = int(input("Enter power (can be negative): "))

p = Power()
print(f"\nResult of {x}^{n} =", p.pow(x, n))
```

```python
class Item:
    def __init__(self, name, description, price):
        self.name = name
        self.description = description
        self.price = price

    def viewFullDescription(self):
        print(f"\nName: {self.name}")
        print(f"Description: {self.description}")
        print(f"Price: {self.price}")

    def addToShoppingBasket(self):
        print(self.name, "added to shopping basket.")

    def removeFromShoppingBasket(self):
        print(self.name, "removed from shopping basket.")


class MP3(Item):
    def __init__(self, name, description, price, artist, duration):
        super().__init__(name, description, price)
        self.artist = artist
        self.duration = duration

    def play(self):
        print(f"Playing {self.name} by {self.artist}")

    def download(self):
        print(f"Downloading {self.name}")


class DVD(Item):
    def __init__(self, name, description, price, certificate, duration, actors):
        super().__init__(name, description, price)
        self.certificate = certificate
        self.duration = duration
        self.actors = actors

    def viewTrailer(self):
        print(f"Viewing trailer of {self.name}")


class Book(Item):
    def __init__(self, name, description, price, author, numberOfPages, genre):
        super().__init__(name, description, price)
        self.author = author
        self.numberOfPages = numberOfPages
        self.genre = genre

    def previewContent(self):
        print(f"Previewing {self.name} by {self.author}")


# ----- User Choice -----
print("Choose item type:")
print("1. MP3")
print("2. DVD")
print("3. Book")
choice = int(input("Enter choice (1-3): "))

if choice == 1:
    name = input("Enter MP3 name: ")
    desc = input("Enter description: ")
    price = float(input("Enter price: "))
    artist = input("Enter artist name: ")
    duration = input("Enter duration: ")
    mp3 = MP3(name, desc, price, artist, duration)
    mp3.viewFullDescription()
    mp3.play()

elif choice == 2:
    name = input("Enter DVD name: ")
    desc = input("Enter description: ")
    price = float(input("Enter price: "))
    cert = input("Enter certificate: ")
    duration = input("Enter duration: ")
    actors = input("Enter actors (comma separated): ").split(",")
    dvd = DVD(name, desc, price, cert, duration, actors)
    dvd.viewFullDescription()
    dvd.viewTrailer()

elif choice == 3:
    name = input("Enter Book name: ")
    desc = input("Enter description: ")
    price = float(input("Enter price: "))
    author = input("Enter author name: ")
    pages = int(input("Enter number of pages: "))
    genre = input("Enter genre: ")
    book = Book(name, desc, price, author, pages, genre)
    book.viewFullDescription()
    book.previewContent()

else:
    print("Invalid choice!")
```
