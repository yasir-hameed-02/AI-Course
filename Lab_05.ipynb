```python
import numpy as np

print("Task 1: Maximum and Minimum Value of Flattened Array")

arr = np.array([[0, 1], [2, 3]])
print("Original flattened array:")
print(arr)

max_val = arr.max()
min_val = arr.min()

print("Maximum value of the above flattened array:", max_val)
print("Minimum value of the above flattened array:", min_val)
```

```python
print("Task 2: 3x3 Matrix with values from 2 to 10")

matrix = np.arange(2, 11).reshape(3, 3)
print(matrix)
```

```python
print("Task 3: Convert Array to Float Type")

arr = np.array([1, 2, 3, 4])
print("Original array:", arr)

float_arr = arr.astype(float)
print("Array converted to float type:", float_arr)
```

```python
print("Task 4: Multiply 2x2 Matrices and Take Transpose")

A = np.array([[5, 5], [1, 1]])
B = np.array([[1, 2], [0, 1]])

result = np.dot(A, B)
print("Resultant Matrix:")
print(result)

transpose = result.T
print("Transpose of Resultant Matrix:")
print(transpose)
```

```python
print("Task 5: Find Inverse of Matrix")

matrix = np.array([
    [6, 1, 1, 3],
    [4, -2, 5, 1],
    [2, 8, 7, 6],
    [3, 1, 9, 7]
])

inverse_matrix = np.linalg.inv(matrix)
print("Inverse of the matrix:")
print(inverse_matrix)
```
