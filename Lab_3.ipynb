```python
def factorial(num):
    fact = 1
    for i in range(1, num + 1):
        fact *= i
    return fact

# Example:
n = int(input("Enter a non-negative integer: "))
print("Factorial =", factorial(n))
```

```python
def unique_list(lst):
    unique = []
    for i in lst:
        if i not in unique:
            unique.append(i)
    return unique

# Example:
sample = [19, 2, 13, 31, 14, 14, 54, 32, 27, 50]
print("Original List:", sample)
print("Unique List:", unique_list(sample))
```

```python
def even_numbers(n):
    evens = []
    for i in range(1, n + 1):
        evens.append(i * 2)
    print("Even numbers:", evens)
    return evens

# Example:
n = int(input("Enter how many even numbers you want: "))
even_list = even_numbers(n)
total_sum = sum(even_list)
product = 1
for num in even_list:
    product *= num
print("Sum =", total_sum)
print("Product =", product)
```

```python
def match_strings(str1, str2):
    if str1 == str2:
        print("Both strings match!")
    else:
        print("Strings do not match.")

# Example:
s1 = input("Enter first string: ")
s2 = input("Enter second string: ")
match_strings(s1, s2)
```
