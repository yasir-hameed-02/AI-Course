```python
import random

rooms = {"A": "Dirty", "B": "Dirty", "C": "Dirty"}
clean_count = 0

while True:
    print("\nCurrent Room States:", rooms)
    current_room = input("Enter current room (A/B/C or 'exit' to stop): ").upper()

    if current_room == 'EXIT':
        break

    if current_room not in rooms:
        print("Invalid room! Choose A, B or C.")
        continue

    if rooms[current_room] == "Dirty":
        print(f"Room {current_room} is dirty. Cleaning now...")
        rooms[current_room] = "Clean"
        clean_count += 1
    else:
        print(f"Room {current_room} is already clean.")

    if all(state == "Clean" for state in rooms.values()):
        print("\n‚úÖ All rooms are clean!")
        random_room = random.choice(list(rooms.keys()))
        print(f"Agent moves randomly to Room {random_room}")
        break

print("\nFinal Room States:", rooms)
print("Total Clean Actions:", clean_count)
```

```python
memory = []

print("Thermostat Active (Range: 20¬∞C to 24¬∞C)")
while True:
    temp = input("\nEnter current temperature (or 'exit' to stop): ")

    if temp.lower() == "exit":
        print("Thermostat turned off.")
        break

    try:
        temp = float(temp)
    except:
        print("Invalid input! Enter a number.")
        continue

    memory.append(temp)
    if len(memory) > 3:
        memory.pop(0)  # keep only last 3 readings

    avg = sum(memory) / len(memory)
    print(f"Recent Temps: {memory} ‚Üí Average = {avg:.1f}¬∞C")

    if avg < 20:
        print("Temperature low ‚Üí Turning ON Heater üî•")
    elif avg > 24:
        print("Temperature high ‚Üí Turning ON Cooler ‚ùÑÔ∏è")
    else:
        print("Temperature Normal ‚Üí Maintaining current state üí§")
```
