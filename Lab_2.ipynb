```python
# New list of strings
words = ['mom', 'hello', 'noon', 'apple', 'dad', 'pop', 'sky', '121']

count = 0
for word in words:
    if len(word) >= 2 and word[0] == word[-1]:
        count += 1

print("Expected Result:", count)
```

```python
numbers = (1, 3, 4, 7, 3)

sum_num = sum(numbers)      # Built-in sum()
product = 1
for n in numbers:
    product *= n

print("Sum is:", sum_num)
print("Product is:", product)
```

```python
numbers = [11, 21, 14, 32, 69]
largest = max(numbers)      # Built-in max() finds largest value
print("Largest number is:", largest)
```

```python
L = [ (), ('Yasir',), ('Khan', 'Baba'), (), ('Barrister', 'Babu', 'Sahab',), (), ('Duplicate',)]

result = [t for t in L if t]   # Keeps only non-empty tuples
print("Result =", result)
```

```python
n = int(input("Enter a number: "))
square_dict = {}

for x in range(1, n + 1):
    square_dict[str(x)] = x * x

print(square_dict)
```

```python
num = input("Enter a number: ")

words = {
    '0': 'Zero', '1': 'One', '2': 'Two', '3': 'Three',
    '4': 'Four', '5': 'Five', '6': 'Six',
    '7': 'Seven', '8': 'Eight', '9': 'Nine'
}

for digit in num:
    print(words[digit], end=' ')
```
